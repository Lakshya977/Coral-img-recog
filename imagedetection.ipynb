```python
import zipfile
import os
zip_file_path = '/content/Coral Bleaching_General.v1i.yolov8.zip'
extract_folder = '/content/coral_bleaching_dataset/'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print("Unzipping completed!")
```

```python
from ultralytics import YOLO

# Load the YOLOv8 Nano model (smallest and fastest)
model = YOLO('yolov8n.pt')


model.train(
    data='/content/coral_bleaching_dataset/data.yaml',
    epochs=5,
    imgsz=416,
    batch=8,
    workers=2,

)
```

```python
from ultralytics import YOLO
model = YOLO("/content/runs/detect/train4/weights/best.pt")
model.val()
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train4/weights/best.pt")

model.predict(source="/content/coral_bleaching_dataset/test/images", save=True)
```
